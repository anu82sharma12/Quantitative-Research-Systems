```python
# EdgeFinance Node — Real-Time Inference Microservice Demo
import time, numpy as np

def predict_price(x):
    # lightweight linear model
    w, b = 1.02, 0.5
    return w*x + b

start = time.time()
for _ in range(1000):
    _ = predict_price(np.random.rand())
latency = (time.time() - start)/1000*1000  # ms
print(f"Inference latency ≈ {latency:.2f} ms (simulated edge run)")
```
