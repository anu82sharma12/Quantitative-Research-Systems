```python
# FinSense Core  Sentiment-Augmented Forecast Model
import numpy as np, matplotlib.pyplot as plt

np.random.seed(0)
days = 300
sentiment = np.random.uniform(-1,1,days)
returns = 0.002*sentiment + np.random.normal(0,0.01,days)
corr = np.corrcoef(sentiment, returns)[0,1]

plt.scatter(sentiment, returns, alpha=0.5)
plt.title(f"FinSense Core  Sentiment vs Return (corr={corr:.2f})")
plt.xlabel("Sentiment"); plt.ylabel("Return")
plt.show()
```
