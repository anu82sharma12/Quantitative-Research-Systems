```python
import gym
import numpy as np
import matplotlib.pyplot as plt

# Synthetic reward simulation
rewards = np.cumsum(np.random.randn(100) * 0.2 + 0.5)
baseline = np.cumsum(np.random.randn(100) * 0.2 + 0.3)

plt.plot(rewards, label="QAlphaX Agent")
plt.plot(baseline, label="Baseline Portfolio")
plt.title("Simulated Portfolio Performance")
plt.legend()
plt.show()
```
