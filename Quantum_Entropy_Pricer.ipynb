```python
# Quantum Entropy Pricer — Monte Carlo Option Simulation
import numpy as np, matplotlib.pyplot as plt

S0,K,r,sigma,T,N = 100,100,0.01,0.2,1,5000
Z = np.random.normal(0,1,N)
ST = S0*np.exp((r-0.5*sigma**2)*T + sigma*np.sqrt(T)*Z)
payoff = np.maximum(ST-K,0)
price = np.exp(-r*T)*np.mean(payoff)
print(f"Option Price ≈ {price:.2f}")

plt.hist(ST, bins=50, alpha=0.7)
plt.title("Quantum Entropy Pricer – Simulated Terminal Prices")
plt.xlabel("Price"); plt.ylabel("Frequency")
plt.show()
```
